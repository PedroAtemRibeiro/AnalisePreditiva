```python

import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score

```

```python
# Importando dados Treino em formato csv para um dataframe
Dados = pd.read_csv('desafio_manutencao_preditiva_treino.csv')
Dados
```

```python
#Importando os dados teste e vinculando dataframe Ã  uma variÃ¡vel e deletando as colunas desnecessÃ¡rias para aplicaÃ§Ã£o do aprendizado de maquina
DadosTeste = pd.read_csv('desafio_manutencao_preditiva_teste.csv')
DadosTeste = DadosTeste.drop(columns=['udi','product_id','type'])
DadosTeste

```

```python
# x recebe as Features // y recebe o predicted
x = Dados.drop(columns=['udi', 'failure_type','product_id','type'])
y = Dados['failure_type']

#importando as variaveis para a funÃ§Ã£o de aplicaÃ§Ã£o de aprendizado de maquina do pacote sklearn e definindo teste como apenas 20%
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,  test_size=0.2)
```

```python
# vindulando uma Variavel Ã  funÃ§Ã£o de arvore de decisÃ£o do pacote sklearn
modelo = DecisionTreeClassifier()

#aplicando a funÃ§Ã£o arvore de decisÃ£o com as variaveis treino e teste, ja que o nosso predict Ã© uma classificaÃ§Ã£o com 1/5 de possibilidades
modelo.fit(x_treino,y_treino)
```

```python
#arvore de decisÃ£o ja estÃ¡ treinada para a classificaÃ§Ã£o

#como Teste aplicaremos a funÃ§Ã£o de arvore de decisÃ£o na variavel x que possui apenas as colunas de features e vinculamos o resultado a uma Variavel 

FullPredictedSheetTrain = pd.DataFrame(modelo.predict(x))
```

```python
#aqui alteramos o nome da coluna para 'PredictedFeedback' para diferenciarmos da coluna 'Failure_type'
FullPredictedSheetTrain.rename(columns = {0:'PredictedFeedback'}, inplace = True)

FullPredictedSheetTrain

# aqui concatenamos o dataframe Dados(com todas as colunas) e o nosso dataframe fruto do aprendizado de maquina e aplicado com os features
TableTrei = pd.concat([Dados,FullPredictedSheetTrain],axis = 1)

#aqui exibimos a tabela concatenada
TableTrei
```

```python

#para medir a taxa de acerto utilizamos a funÃ§Ã£o 'accuracy' do pacote sklearn para comparar os valores de Y com  os da coluna 'PredictedFeedback' obtidos com a arvore de decisÃ£o e vinculamos a uma variavel 'taxa_acerto'
taxa_acerto = accuracy_score(y,FullPredictedSheetTrain)

#valor de taxa de acerto em porcentagem, ou seja, 99% de acertos
print(taxa_acerto)
```

```python
#exportamos para um csv criado para comparaÃ§Ã£o e visualizaÃ§Ã£o
TableTrei.to_csv('Feedback.csv')
# Caminho do arquivo --^ (Se nÃ£o estiver dentro do Folder)
```
