```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
treino = pd.read_csv('Treino.csv', sep=',')
```

```python
teste = pd.read_csv('Teste.csv', sep=',')
```

```python
treino
```

```python
teste

```

```python
labelencoder = LabelEncoder()
for i in treino.select_dtypes(include ='object'):
        treino[i] = labelencoder.fit_transform(treino[i])
```

```python
labelencoder = LabelEncoder()
for i in teste.select_dtypes(include ='object'):
        teste[i] = labelencoder.fit_transform(teste[i])
```

```python
teste.rename(columns = {'udi': 'RowNumber'}, inplace = True)
```

```python
teste
```

```python
treino.head()
```

```python
x = pd.DataFrame(treino.iloc[:,-2])
```

```python
y = pd.DataFrame(treino.iloc[:,-1])
```

```python
y_test = pd.DataFrame(teste.iloc[:,-1])
```

```python
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
print(x_treinamento.shape)
print(x_teste.shape)
print(y_treinamento.shape)
print(y_teste.shape)
```

```python
floresta = RandomForestClassifier(n_estimators = 100, random_state=0)
floresta.fit(x_treinamento, y_treinamento)
```

```python
previsoes = floresta.predict(x_teste)

print(previsoes)
```

```python
confusao = confusion_matrix(y_teste,previsoes)
print(confusao)
```

```python
taxa_acerto = accuracy_score(y_teste,previsoes)
print(taxa_acerto)
```

```python
predict = floresta.predict(y_test)
```

```python
predict
```

```python
pred = pd.DataFrame(predict)
```

```python
pred.rename(columns = {0:'Predicted'}, inplace = True)
```

```python
pred
```

```python
row_num = pd.DataFrame(teste.iloc[:,0:1])
```

```python
pd.concat([row_num,pred],axis = 1)
```

```python
Tabela = pd.concat([row_num,pred],axis = 1)
```

```python
Tabela
```

```python
labelencoder = LabelEncoder()
for i in pred.select_dtypes(include ='int'):
        pred[i] = labelencoder.fit_transform(pred[i])
```

```python
Tabela = Tabela.set_index('RowNumber')
```
